```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
```

```python
from api_keys import api_key
```

```python
from citipy import citipy
```

```python
output_data_file = "output_data/cities.csv"
```

```python
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
lat_lngs = []
cities = []
```

```python
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
```

```python
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)
```

```python
len(cities)
```

```python
base_url= "http://api.openweathermap.org/data/2.5/weather?&APPID=" + api_key 
```

```python
city_name=[]
cloudiness=[]
country=[]
date=[]
humidity=[]
lat=[]
lng=[]
temp=[]
wind_speed=[]
```

```python
record =1
```

```python
for city in cities:
    try:
        response=requests.get(f"{base_url}&q={city}").json()
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        temp.append(response["main"]["temp_max"])
        wind_speed.append(response["wind"]["speed"])
        record=response["name"]    
        print(f" Beginning Data Retrieval")
        print(f"-------------------------")
        print(f"Processing Record {record} |{city_name}")
        record=record+1
        time.sleep(1.01)
    except(KeyError, TypeError):
          print("City Not Found. Skipping")
```

```python
weather_dictionary={"City Name": city_name, "Cloudiness": cloudiness, "Country":country, "Date": date, "Humidity": humidity, "Lat": lat, "Lng": lng, "Max Temp":temp,"Wind Speed": wind_speed}
```

```python
weather_data=pd.DataFrame(weather_dictionary)
```

```python
weather_data.head()
```
